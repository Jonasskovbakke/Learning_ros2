```python
import matplotlib.pyplot as plt
import numpy as np
import ipympl

#               x    y    z
p = np.array([[[0], [0], [0]],
              [[0], [0], [1]]])

#                 x    y    z
p_m = np.array([[[1], [0], [0]],
                [[2], [0], [0]]])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(p[0, 0, 0], p[0, 1, 0], p[0, 2, 0], color="red")
ax.scatter(p[1, 0, 0], p[1, 1, 0], p[1, 2, 0], color="red")
ax.scatter(p_m[0, 0, 0], p_m[0, 1, 0], p_m[0, 2, 0], color="green")
ax.scatter(p_m[1, 0, 0], p_m[1, 1, 0], p_m[1, 2, 0], color="green")


ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_zlabel("z-axis")

plt.show()
```

```python
c_p = sum(p) / len(p)
print(c_p)

c_p_m = sum(p_m) / len(p_m)
print(c_p_m)
```

```python
q = p - c_p
q_m = p_m - c_p_m

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(q[0, 0, 0], q[0, 1, 0], q[0, 2, 0], color="red")
ax.scatter(q[1, 0, 0], q[1, 1, 0], q[1, 2, 0], color="red")
ax.scatter(q_m[0, 0, 0], q_m[0, 1, 0], q_m[0, 2, 0], color="green")
ax.scatter(q_m[1, 0, 0], q_m[1, 1, 0], q_m[1, 2, 0], color="green")

ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_zlabel("z-axis")

plt.show()
```

```python
H = np.zeros([3, 3])
for i in range(len(q)):
    H += q[i] @ q_m[i].T

print(H)
```

```python
u, s, vh = np.linalg.svd(H, full_matrices=True)
R = vh @ u.T
print(R)
```

```python
T = c_p_m - R @ c_p
print(T)
```

```python
H_rb = np.eye(4)
print(H_rb[0:3, 0:3])
H_rb[0:3, 0:3] = R
print(H_rb)
H_rb[0:3, 3] = T[0:3].T[0]
print(H_rb)
```
