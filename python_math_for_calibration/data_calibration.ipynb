```python
read_x = data_x.flatten()
#print(read_x)

read_y = data_y.T.flatten()
#print(read_y)

matrix = np.empty([len(read_x), 2, 1])

for i in range(len(read_x)):
    vector = np.array([[read_x[i]], [read_y[i]]])
    matrix[i] = vector

print(matrix)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import ipympl


fig = plt.figure()
""" ax = fig.add_subplot(projection='3d') """
ax = fig.add_subplot()

for i in range(len(matrix)):
    ax.scatter(matrix[i, 0, 0], matrix[i, 1, 0], color="red")

ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

plt.show()
```
